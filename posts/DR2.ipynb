### 使用`purrr`实现迭代

#### `for` 循环

```r
df <- tibble(
    a= rnorm(10),
    b= rnorm(10),
    c= rnorm(10),
    d = rnorm(10)
)
```

```r
df
```

```r
c(mean(df$a),mean(df$b))
```

每个循环都包含三个部分。

输出：`output<-vector("double",length(x))`

开始循环时候，必须为输出结果分配足够的时间，对循环效率来说比较重要，若在每次迭代时候，都是用`c()`来保存循环结果，`for`循环就非常慢

序列：`i in seq_along(df)`

循环体： `output[[i]]<-median(df[[i]])`


```r
rescale01<-function(x){
    rng<-range(x,na.rm=TRUE)
    (x-rng[1])/(rng[2]-rng[1])
}
```

```r
df$a<-rescale01(df$a)
```

```r
df$a
```

```r
for (i in seq_along(df)){
    df[[i]]<-rescale01(df[[i]])
}
```

**note**:循环列表或数据框，使用`[[`

循环的方式：
- 元素循环
- 名称循环


使用索引值进行循环是最常用的方式



#### 未知长度的输出

```r
means <-c(0,1,2)
output<-double()
```

```r
for (i in seq_along(means)){
    n <- sample(100,1)
    output<- c(output,rnorm(n,means[[i]]))
}
str(output)
```

#### for 与函数编程

```r
df <- tibble(
    a= rnorm(10),
    b= rnorm(10),
    c= rnorm(10),
    d= rnorm(10)
)
```

```r
output <- vector("double",length(df))
```

```r
for (i in seq_along(df)){
    output[[i]]=mean(df[[i]])
}
```

```r
output
```

```r
col_summary<-function(df,fun){
    out<-vector("double",length(df))
    for (i in seq_along(df)){
        out[i]<-fun(df[[i]])
    }
    out
}
```

```r
col_summary(df,median)
```

```r
col_summary(df,mean)
```

将函数作为参数传入。

## 模型

一个简单的模型

```r
library(modelr)
options(na.action=na.warn)
```

```r
ggplot(sim1,aes(x,y))+
    geom_point()
```

使用`geom_abline()`作为接受斜率和截距的参数。

```r
models<-tibble(
    a1=runif(250,-20,40),
    a2=runif(250,-5,5)
)
```

```r
ggplot(sim1,aes(x,y))+
    geom_abline(
        aes(intercept=a1,slope=a2),
        data=models,alpha=1/4)+
    geom_point()
```

```r
model1<-function(a,data){
    a[1]+data$x*a[2]
}
model1(c(7,1.5),sim1)
```

```r
lm(sim1$y~sim1$x)
```

计算RMSE

```r
measure_distance<-function(mod,data){
    diff<-data$y-model1(mod,data)
    sqrt(mean(diff^2))

measure_distance(c(7,1.5),sim1)
```

```r
ggplot(sim1,aes(x,y))+
    geom_abline(
        aes(intercept=4.221,slope=2.052),
        data=models,alpha=1/4)+
    geom_point()
```

```r
sim1_dist<-function(a1,a2){
    measure_distance(c(a1,a2),sim1)
}
models<- models%>%
    mutate(dist=purrr::map2_dbl(a1,a2,sim1_dist))
models
```

```r
ggplot(sim1,aes(x,y))+
    geom_point(size=2,color="grey30")+
    geom_abline(
        aes(intercept=a1,slope=a2,color=-dist),
        data=dplyr::filter(models,rank(dist)<=10)
    )
```

```r
best<-optim(c(0,0),measure_distance,data=sim1)
best$par
```

```r
lm(sim1$y~sim1$x)
```

```r
ggplot(sim1,aes(x,y))+
    geom_abline(
        aes(intercept=best$par[1],slope=best$par[2]),
        data=models,alpha=1/4)+
    geom_point()
```
